```python
Q1 = df2["Amount"].quantile(0.25)
Q3 = df2["Amount"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
```

```python
df_price_outliers = df2[(df2["Amount"] > upper_bound) | (df2["Amount"] < lower_bound)]
print(round(((len(df_price_outliers.index)/len(df2.index))*100), 2))
df_price_outliers
```

# Analysis

```python
df_sales = df2.copy()
```

```python
df_sales[["Month", "Day", "Year"]] = df_sales["Date"].str.split("-", expand=True)
```

```python
df_sales["ship-state"] = df_sales["ship-state"].str.title()
```

```python
df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "New Delhi" if x == "Delhi" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Puducherry" if x == "Pondicherry" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Punjab" if x == "Pb" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Punjab" if x == "Punjab/Mohali/Zirakpur" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Nagaland" if x == "Nl" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Rajasthan" if x == "Rj" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Rajasthan" if x == "Rajsthan" else x
)

df_sales["ship-state"] = df_sales["ship-state"].apply(
    lambda x: "Arunachal Pradesh" if x == "Ar" else x
)
```

```python
df_sales["ship-state"].unique()
```

### Monthly Sales

```python
df_monthly_sales = df_sales.groupby("Month").agg(
    monthly_sales = ("Amount", "sum")
)
# df_monthly_sales.reset_index(inplace=True)
df_monthly_sales
```

```python
df_monthly_sales.plot(kind="line")
```

### State-wise Sale

```python
df_state_sales = df_sales.groupby("ship-state").agg(
    no_order = ("Order ID", "nunique"),
    sales = ("Amount", "sum")
)
df_state_sales.sort_values(by="sales", ascending=False, inplace=True)
df_state_sales.reset_index(inplace=True)
df_state_sales.head(10)
```

```python
top_ten_states_by_sales = df_state_sales.head(10)

plt.figure(figsize=(20, 12))

plt.subplot(2,1,1)
plt.bar(top_ten_states_by_sales["ship-state"], top_ten_states_by_sales["sales"])
plt.title("State-vs-Sale (Top-10)")

plt.subplot(2,1,2)
plt.bar(top_ten_states_by_sales["ship-state"], top_ten_states_by_sales["no_order"])
plt.title("State-vs-Orders (Top-10)")

plt.tight_layout()
plt.show()
```

### Category Analysis

```python
df_sales["Category"].unique()
```

```python
df_month_category = df_sales.groupby(["Month", "Category"]).agg(
    no_orders = ("Order ID", "nunique") ,
    monthly_sales = ("Amount", "sum")
)
df_month_category
```

```python
df_set_monthly_sales = df_month_category.xs("Set", level="Category")
df_kurta_monthly_sales = df_month_category.xs("kurta", level="Category")
df_wes_monthly_sales = df_month_category.xs("Western Dress", level="Category")
df_top_monthly_sales = df_month_category.xs("Top", level="Category")
df_eth_monthly_sales = df_month_category.xs("Ethnic Dress", level="Category")
df_bot_monthly_sales = df_month_category.xs("Bottom", level="Category")
df_sar_monthly_sales = df_month_category.xs("Saree", level="Category")
df_blo_monthly_sales = df_month_category.xs("Blouse", level="Category")
df_dup_monthly_sales = df_month_category.xs("Dupatta", level="Category")

plt.figure(figsize=(15,9))
plt.subplot(3,3,1)
df_set_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Set Monthly Sales")

plt.subplot(3,3,2)
df_kurta_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Kurta Monthly Sales")

plt.subplot(3,3,3)
df_wes_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Western Dress Monthly Sales")

plt.subplot(3,3,4)
df_top_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Top Monthly Sales")

plt.subplot(3,3,5)
df_eth_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Ethnic Dress Monthly Sales")

plt.subplot(3,3,6)
df_bot_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Bottom Monthly Sales")

plt.subplot(3,3,7)
df_sar_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Saree Monthly Sales")

plt.subplot(3,3,8)
df_blo_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Blouse Monthly Sales")

plt.subplot(3,3,9)
df_dup_monthly_sales["monthly_sales"].plot(kind="line")
plt.title("Dupatta Monthly Sales")


plt.tight_layout()
plt.show()
```
